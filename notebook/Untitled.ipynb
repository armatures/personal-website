```python
from pandas import read_csv
```

```python
s = read_csv('Telemarketing_RoboCall_Weekly_Data.csv')
s
```

```python
s['Time of Issue'].fillna(s['Date (Ticket Date of Issue)'], inplace=True)
s.loc[:,['Time of Issue']]
```

```python
from datetime import datetime
def parse_time(raw_time):
    if raw_time == '-':
        return None
    time = raw_time.upper().replace(".", "").zfill(8)
    print(time)
    return datetime.strptime(
        time,
        "%I:%M %p",
    )

for t in s.loc[:,['Time of Issue', 'Date (Ticket Date of Issue)']]:
    print(t)
    print(t, parse_time(t))
```
